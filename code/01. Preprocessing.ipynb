```python
import pandas as pd
import numpy as np
```

# Data Load & Preprocessing

```python
raw_data = pd.read_csv('table_lv7_rf_silva.tsv',sep='\t')
raw_data
```

```python
# 행과 열이 제대로 분리되어있지않은 데이터프레임을 사전 자료형을 이용해 분리

df_dict = {}
for i in range(len(raw_data.T.columns)):
    tmp_sr = raw_data.T.columns[i]
    df_dict[tmp_sr[0]] = np.array(tmp_sr[1:])
df = pd.DataFrame(df_dict).set_index('#OTU ID').T.reset_index(drop=False)
df
```

```python
# genus level의 박테리아(?)를 추출해내기 위한 새로운 함수 정의

def return_genus(col):
    for i in col.split(';'):
        if i.startswith('g__'):
            return i
    return col
df['index'] = df['index'].map(return_genus)
df
```

```python
# 같은 genus level 끼리 묶은 후 genus level이 없다면 삭제

df = df.groupby('index').sum()
df = df[df.index.str.contains('g__')]
df
```

```python
df.to_csv('df_genus_3.csv')
```
